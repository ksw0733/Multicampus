```python
def AND_gate(x1, x2):
    weight = np.array([1,1])
    bias = -1
    x = np.array([x1,x2])
    sum = np.dot(weight, x)+ bias
    return 0 if sum <=0 else 1
```

```python
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]:
    print(AND_gate(x1,x2))
```

- OR Gate

```python
def OR_gate(x1, x2):
    weight = np.array([2,2])
    bias = -1
    x = np.array([x1,x2])
    sum = np.dot(weight, x)+ bias
    return 0 if sum <=0 else 1
```

```python
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]:
    print(OR_gate(x1,x2))
```

- NAND Gate

```python
def NAND_gate(x1, x2):
    weight = np.array([-2,-2])
    bias = 3
    x = np.array([x1,x2])
    sum = np.dot(weight, x)+ bias
    return 0 if sum <=0 else 1
```

```python
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]:
    print(NAND_gate(x1,x2))
```

- MLP

```python
n1, n2 = [], []
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]:
    n1.append(NAND_gate(x1,x2))
    n2.append(OR_gate(x1,x2))
```

```python
n1, n2
```

```python
for x1, x2 in zip(n1,n2):
    print(AND_gate(x1, x2))
```

- XOR gate

```python
def XOR_gate(x1, x2):
    return AND_gate(NAND_gate(x1, x2), OR_gate(x1, x2))
```

```python
print('x1,x2\tn1,n2\tXOR')
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]:
    print(f'{x1:2d},{x2:2d}\t{NAND_gate(x1,x2):2d},{OR_gate(x1,x2):2d}\t{XOR_gate(x1,x2):3d}')
```
